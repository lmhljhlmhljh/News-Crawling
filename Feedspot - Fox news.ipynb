```python
# 스크롤하기
SCROLL_PAUSE_TIME = 1.5

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
a = soup.find_all(class_ = 'ext_link')
```

```python
# Feedspot에서 href link list화
business = re.findall('href\=.(.*?)..rel',str(a))
business = list(set(business))
len(business)
```

```python
with open('ABC.csv', 'w', encoding = 'UTF-8', newline= '') as f:
    writer = csv.writer(f)
    writer.writerow(['title', 'content', 'date', 'section', 'company', 'url'])
    
    num = 0
    for i in business :
        
        try :
            html = urlopen(i)
            soup = BeautifulSoup(html, "html.parser")

           #topic
            soup_title = soup.find(class_ = 'article-header')
            soup_title = soup_title.find('h1')
            title = soup_title.get_text()
            title = title.strip()

            #Content
            c = soup.find_all('p')

            list_ = []
            for i in c :
                apple = i.get_text()
                apple = re.sub('\n','',apple)
            list_.append(apple) 
            content = ' '.join(list_)
            print(content)

            #Date
            date = 'Unknown'
    
            #Section
            section ='1'
    
            #신문사
            company = 'Forbes'
        
            #url
            url = a
        
        except :
            print('{}실패!'.format(num))
            
        writer.writerow([title, content, date, section, company, url]) 
```
