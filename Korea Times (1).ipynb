```python
# Import module
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys # 스크롤
import csv
import time
import codecs
import nltk, re
from selenium.webdriver.common.alert import Alert
from urllib.request import urlopen
```

```python
# 1. Selenium module 가져오기
driver = webdriver.Chrome('/Users/82107/chromedriver')
driver.get('https://www.koreatimes.co.kr/www/sublist_398.html')
```

```python
html_list = []
```

```python
# 2. Get Url
html = driver.page_source
Ktimes = BeautifulSoup(html, 'html.parser')

Ktimes_html = re.findall('\/www\/art.*?\.html', str(Ktimes))
Ktimes_html = list(set(Ktimes_html))
for word in Ktimes_html :
    word_ = "https://www.koreatimes.co.kr" + word
    html_list.append(word_)
html_list
```

```python
# 스크롤하기
SCROLL_PAUSE_TIME = 0.5

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
# 3. Crawling information from 1 page.
a = 'https://www.koreatimes.co.kr/www/nation/2019/09/120_274096.html'
html = urlopen(a)
k_times = BeautifulSoup(html, "html.parser")

headline = k_times.find('div', class_ ='view_headline')
headline = headline.get_text()
print(headline)

news = k_times.find('div', {'id' : 'startts'})
news = news.get_text()
print(news)
```

```python
# URL list 가져오기
html_list_2 = []
```

```python
# 4. Get URL from 100 pages 
num = 0
while num < 300 :
    num += 1
    
    # 1~9페이지 URL 
    if num % 10 != 9 :
        #Scroll
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        html = driver.page_source
        Ktimes = BeautifulSoup(html, 'html.parser')
        
        #URL 편집
        Ktimes_html = re.findall('\/www\/art.*?\.html', str(Ktimes))
        Ktimes_html = list(set(Ktimes_html))
        for word in Ktimes_html :
            word_ = "https://www.koreatimes.co.kr" + word
            html_list_2.append(word_)
        driver.find_element_by_xpath('/html/body/div[6]/div[2]/div[1]/div[12]/span[2]/a[{}]'.format(num%10+1)).click()
    
    # 10페이지 URL 
    else :
        #Scroll
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        html = driver.page_source
        Ktimes = BeautifulSoup(html, 'html.parser')

        #URL 편집
        Ktimes_html = re.findall('\/www\/nation.*?\.html', str(Ktimes))
        Ktimes_html = list(set(Ktimes_html))
        for word in Ktimes_html :
            word_ = "https://www.koreatimes.co.kr" + word
            html_list_2.append(word_)
        driver.find_element_by_xpath('/html/body/div[6]/div[2]/div[1]/div[12]/span[3]/a/img[1]').click()
```

```python
# Url 주소 저장
import json
with open('html_list(7000), 1200p.json', 'w') as f:
    json.dump(html_list_2, f)
f.close()
```

```python
# Url 주소 불러오기
import json
with open('html_list(), 1200page.json') as json_file:
    html_list = json.load(json_file)
```

```python
# 5.get the information from several pages
num = 2932
with open('Korean_times_entertainment.csv', 'a', encoding = 'utf-8', newline= '') as f:
    writer = csv.writer(f)
    for saved_html in html_list_2 :
        html = urlopen(saved_html)
        k_times = BeautifulSoup(html, "html.parser")

        headline = k_times.find('div', class_ ='view_headline')
        headline = headline.get_text()

        news = k_times.find('div', {'id' : 'startts'})
        news = news.get_text()
        writer.writerow([headline, news])
        num += 1
        print('{}번 성공'.format(num))
```
