```python
# 2.0 Crawling from one page
html = urlopen('http://www.koreaherald.com/list.php?ct=020100000000&ctv=0&np=1')
soup = BeautifulSoup(html, 'html.parser')
url_soup = soup.findAll('li', class_ = 'mb24 pl55')
url_result = re.findall('href\=\"(.*?[0-9]{14})\"',str(url_soup))
url_result =list(set(url_result))
print(url_result)
print(len(url_result))
```

```python
# 2. URL 가져오기 by using Beautifulsoup
start_num = 1
final_num = 1100
start = time.time() 

for num in range(start_num, final_num) :
    html = urlopen('http://www.koreaherald.com/list.php?ct=020600000000&ctv=0&np={}'.format(num))
    soup = BeautifulSoup(html, 'html.parser')
    url_soup = soup.findAll('li', class_ = 'mb24 pl55')
    url_result = re.findall('href\=\"(.*?[0-9]{14})\"',str(url_soup))
    url_result =list(set(url_result))
    KHerald_opinion.extend(url_result)
    msg = '\r{}번 성공'.format(num)
    print(msg, end='')
    continue

KHerald_opinion = list(set(KHerald_opinion))
duration = "time :{}".format(time.time() - start)
print(msg)
print(duration)
print(len(KHerald_opinion))
```

```python
start_num = 1
final_num = len(KHerald_opinion)
num = start_num
start = time.time() 

with open('KHerald_opinion.csv', 'a', encoding = 'utf-8', newline= '') as f:
    writer = csv.writer(f)
    #writer.writerow(['headline', 'content'])
    
    for saved_html in KHerald_opinion[start_num:final_num] :
        url = 'http://www.koreaherald.com' + saved_html 
        html = urlopen(url)
        soup = BeautifulSoup(html, "html.parser")

        #headline
        headline = soup.find('p', class_ = 'fontTitle6 mb16')
        headline = headline.get_text()

        #content
        content = soup.find('div', class_ = 'content_view')
        content = content.get_text()
        writer.writerow([headline, content])
        
        #count numbers
        num += 1
        msg = '\r{}번 성공'.format(num)
        print(msg, end='')

duration = "time :{}".format(time.time() - start)
print(msg, duration)
```
