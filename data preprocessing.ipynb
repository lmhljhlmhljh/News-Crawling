```python
# Open CSV
df = pd.read_csv('2nd_Week(220413).csv', encoding='utf-8', engine = 'python')
```

## 전처리(Preprocessing)

```python
#0. making data to str
def make_it_data(row) : 
    # 2칸 이상의 공백 / 의미 없는 줄바꿈 / data 형식 표시 제거
    ex_ = re.sub(' {2,}|\n|Name.*?object',' ',str(row))
    # 숫자 제거
    ex_ = re.sub('[0-9]',' ',ex_)
    # 앞 뒤 공백 제거
    ex_ = ex_.strip()
    return ex_
```

```python
#2. 특수문자 (special character)
def sc_del(row) :
    row_ = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]',' ',row)
    return row_
```

```python
# apply functions to remove numbers and special character
df.title = df.title.apply(make_it_data)
df.title = df.title.apply(sc_del)
df.content = df.content.apply(make_it_data)
df.contnet = df.content.apply(sc_del)
df
```

```python
df.to_csv('2nd_Week_edit(220413).csv', index=False, encoding= 'utf-8')
```
