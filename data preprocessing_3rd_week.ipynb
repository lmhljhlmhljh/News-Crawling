## 전처리(Preprocessing)

```python
#0. making data to str
def make_it_data(row) : 
    # 2칸 이상의 공백 / 의미 없는 줄바꿈 / data 형식 표시 제거
    ex_ = re.sub(' {2,}|\n|Name.*?object',' ',str(row))
    # 숫자 제거
    ex_ = re.sub('[0-9]',' ',ex_)
    # 앞 뒤 공백 제거
    ex_ = ex_.strip()
    return ex_
```

```python
#2. 특수문자 (special character)
def sc_del(row) :
    row_ = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]',' ',row)
    return row_
```

```python
# apply functions to remove numbers and special character
df.title = df.title.apply(lambda x : make_it_data(x))
df.title = df.title.apply(lambda x : sc_del(x))
df.content = df.content.apply(lambda x : make_it_data(x))
df.contnet = df.content.apply(lambda x : sc_del(x))
df
```

```python
# 전처리

df.content = df.content.apply(lambda x : re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]','',str(x)))
df.title = df.title.apply(lambda x : re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]','',str(x)))
```

```python
df.title = df.title.apply(lambda x : re.sub('[^A-Za-z| ]',' ',str(x)))
df.content = df.content.apply(lambda x : re.sub('[^A-Za-z| ] ',' ',str(x)))
df
```
