```python
# 2. URL 가져오기 by using Beautifulsoup
start_num = 1
final_num = 1455
num_ = start_num

for num in range(start_num, final_num) :
    html = urlopen('http://koreajoongangdaily.joins.com/news/list/List.aspx?gCat=010101&pgi={}'.format(num))
    soup = BeautifulSoup(html, 'html.parser')
    url_soup = soup.findAll('a',class_='title_cr')
    url_result = re.findall('href\=\"(\/news.*?[0-9]{7})\"',str(url_soup))
    html_editorial.extend(url_result)
    print('{}번 성공'.format(num_))
    num_ += 1
    continue
```

```python
# Crawling from one page (중앙데일리)

html = urlopen('http://koreajoongangdaily.joins.com/news/article/Article.aspx?aid=3067881')
soup = BeautifulSoup(html, 'html.parser')

#headline
soup_headline = soup.find('h3',{'id':'sTitle_a'})
headline = soup_headline.get_text()
print(headline)

#content
soup_content = soup.find('div',{'id':'articlebody'})
content = soup_content.get_text()
content = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'\t…》]|\n',' ',content)
content = re.sub('BY.*?kr','',content)
content = content.strip()
print(content)
```

```python
start_num = 10210
final_num = len(html_editorial)
num = start_num

with open('joongang_daily_editorial.csv', 'a', encoding = 'utf-8', newline= '') as f:
    writer = csv.writer(f)
    
    for saved_html in html_editorial[start_num:final_num] :
        url = 'http://koreajoongangdaily.joins.com' + saved_html 
        html = urlopen(url)
        soup = BeautifulSoup(html, "html.parser")

        #headline
        soup_headline = soup.find('h3',{'id':'sTitle_a'})
        headline = soup_headline.get_text()

        #content
        soup_content = soup.find('div',{'id':'articlebody'})
        content = soup_content.get_text()
        content = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'\t…》]|\n',' ',content)
        content = re.sub('BY.*?kr','',content)
        content = content.strip()
        writer.writerow([headline, content])
        num += 1
        print('{}성공'.format(num))
```
